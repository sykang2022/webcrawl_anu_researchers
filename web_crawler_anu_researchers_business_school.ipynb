```python
def expert(x):
    res=requests.get(link[x])
    html=res.text
    soup=BeautifulSoup(html,'html.parser')
    if soup.find(name='div',class_='section',id='areas_of_expertise')== None:
        return 'NNN/AAA'
    else:
        item_list = soup.find(name='div',class_='section',id='areas_of_expertise').find_all(name='li', attrs={'class':'area_of_expertise'}) 
        expertise=list()
        for item in item_list:
            a = item.find(name='span', attrs={'class':'name'}) 
            a=a.text.strip()
            expertise.append(a)
        return str(expertise)
```

```python
def interests(x):
    res=requests.get(link[x])
    html=res.text
    soup=BeautifulSoup(html,'html.parser')
    if soup.find(name='div',class_='section',id='research_interests')== None:
        return 'N/A' 
    else:
        item_list_2 = soup.find(name='div',class_='section',id='research_interests').find_all(name='li') 
        item_list_3 = soup.find(name='div',class_='section',id='research_interests').find_all(name='p') 
        b=item_list_2+item_list_3
        d=str(b)
    
        d=d.replace('<li>','')
        d=d.replace('</li','')
        d=d.replace('</p>','')
        d=d.replace('<p>','')
        d=d.replace('[','')
        d=d.replace(']','')
        d=d.replace('>','')
        d=d.replace('\xa0','')
        d=d.replace('</strong','')
        d=d.replace('<strong','')
        d=d.replace('<em','')
        d=d.replace('</em','')
        d=d.replace('<span','')
        d=d.replace('<p class="western"Please see<a href="https://francishui.netlify.com/"https://francishui.netlify.com/</afor my personal website, CV, and research interests., ','')

    return d
```
